```python
mobilenet_v2 = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
```

```python
mobile_net_layers = hub.KerasLayer(mobilenet_v2, input_shape=(224,224,3))
```

```python
mobile_net_layers.trainable = False
```

```python
neural_net = tf.keras.Sequential([
  mobile_net_layers,
  tf.keras.layers.Dropout(0.3),
  tf.keras.layers.Dense(3,activation='softmax')])
```

Ovdje pazi ovaj rescale faktor - mislim da nam je on suvisan ako su sve slike u png

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_data_dir =project_path+'/training_dataset/train/'

# Set up data augmentation and preprocessing for training dataset
train_data_gen = ImageDataGenerator(
    #rescale=1./255, - ovo brisem ako cemo sve u png korisitit, ako cemo jpeg, treba ostaviti 
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # Splitting the data into train/validation sets
)

# Generate the training dataset
batch_size = 16
image_size = (224, 224)

train_generator = train_data_gen.flow_from_directory(
    train_data_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'  # Use the subset 'training' for training data
)

# Generate the validation dataset
val_generator = train_data_gen.flow_from_directory(
    train_data_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'  # Use the subset 'validation' for validation data
)
```

```python
batch_size = 32
test_data_dir=project_path+'/training_dataset/test/'

test_data_gen = ImageDataGenerator()#rescale=1./255)

# Generate the test dataset
batch_size = 32
image_size = (224, 224)

test_generator = test_data_gen.flow_from_directory(
    test_data_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
   # Use the subset 'validation' for validation data
    
    )  # Use the subset 'training' for training data
```

```python
#@duje mislim da nam je ovdje dobro sto je  categoricalcross jer su u rijeicma


neural_net.compile(optimizer=tf.keras.optimizers.Adam(),loss=tf.keras.losses.CategoricalCrossentropy(),metrics=['accuracy'])
#learning_rate=0.0001)
#neural_net.compile(optimizer=tf.keras.optimizers.Adam(),loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
```

```python
model_fit = neural_net.fit(train_generator, epochs=6, validation_data=val_generator)      
```

```python
acc = model_fit.history['accuracy']
val_acc = model_fit.history['val_accuracy']

loss = model_fit.history['loss']
val_loss = model_fit.history['val_loss']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.ylim([0,1.0])
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

```python
#finetuning
```

```python
mobile_net_layers.trainable = True
layers_to_freeze=range(30)
layers=mobile_net_layers._flatten_layers(recursive=False)
```

```python
for layer in layers:
    if layer.name.split("/")[0] in layers_to_freeze:
        layer.trainable=False
    else:
        layer.trainable=True
```

```python
neural_net = tf.keras.Sequential([
  mobile_net_layers,
  tf.keras.layers.Dropout(0.3),
  tf.keras.layers.Dense(3,activation='softmax')
])

neural_net.compile(
  optimizer=tf.keras.optimizers.Adam(),
  loss=tf.keras.losses.CategoricalCrossentropy(),
  metrics=['accuracy'])
#learning_rate=0.0001)
```

```python
model_fit = neural_net.fit(train_generator, epochs=6, validation_data=val_generator)  
```

```python
acc = model_fit.history['accuracy']
val_acc = model_fit.history['val_accuracy']

loss = model_fit.history['loss']
val_loss = model_fit.history['val_loss']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.ylim([0,1.0])
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```
